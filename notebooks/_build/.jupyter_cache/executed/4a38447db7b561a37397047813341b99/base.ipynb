```python
import numpy as np
import matplotlib.pyplot as plt

loc = 2
scale = 3
size = 10000

data = np.random.normal(loc=loc, scale=scale, size=size)
x = np.linspace(-20, 20, size)

fig, ax = plt.subplots()
ax.hist(data, bins=100)
ax.set_xlim(-20, 20)
plt.show()
```

```python
z = (data - np.mean(data)) / np.std(data)
fig, ax = plt.subplots()
ax.hist(z, bins=100)
ax.set_xlim(-20, 20)
plt.show()
```
